###KPI

```python
total_revenue = df['total_price'].sum()
total_pizzas_sold = df['quantity'].sum()
total_orders= df['order_id'].nunique()

avg_order_value = total_revenue / total_orders
avg_pizzas_per_order = total_pizzas_sold / total_orders

print("Total_revenue : ", total_revenue)
print("Total_pizzas_sold : ", total_pizzas_sold)
print('total_orders :',total_orders)
print(f"avg_order_value is : {avg_order_value:.2f}")
print(f"avg_pizzas_per_order is : {avg_pizzas_per_order:.2f}")
```

```python
##Ingreditents Analysis

ingredients = df['pizza_ingredients'].str.split(',').explode().str.strip().value_counts().reset_index()
print(ingredients)
```

```python
## Daily Trend - total orders

df['order_date'] = pd.to_datetime(df['order_date'],dayfirst = True)
df['day_name'] = df['order_date'].dt.day_name()

weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

df['day_name'] = pd.Categorical(df['day_name'],categories=weekday_order,ordered=True)

order_by_day = df.groupby('day_name',observed=False)['order_id'].nunique()

ax = order_by_day.plot(kind='bar', color='pink', figsize=(10,6), edgecolor='black')

plt.title("Total Orders by Week day", fontsize = 20)
plt.xlabel("Day of the Week", fontsize = 14)
plt.ylabel("Number of Orders", fontsize = 14)

for i,j in enumerate(order_by_day):
    plt.text(i, j+1, str(j), ha='center', fontsize=12, va = 'bottom', fontstyle='italic')
plt.tight_layout()
plt.show()
```

```python
## Daily Trend - total revenue

df['order_date'] = pd.to_datetime(df['order_date'],dayfirst = True)
df['day_name'] = df['order_date'].dt.day_name()

weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

df['day_name'] = pd.Categorical(df['day_name'],categories=weekday_order,ordered=True)

order_by_day = df.groupby('day_name',observed=False)['total_price'].sum()

ax = order_by_day.plot(kind='bar', color='green', figsize=(10,6), edgecolor='black')

plt.title("Total revenue by Week day", fontsize = 20)
plt.xlabel("Day of the Week", fontsize = 14)
plt.ylabel("Total amount", fontsize = 14)

for i,j in enumerate(order_by_day):
    plt.text(i, j+1, f'${j:,.2f}', ha='center', fontsize=12, va = 'bottom', fontstyle='italic')
plt.tight_layout()
plt.show()
```

```python
#Hourly Trend - total orders

df['order_time'] = pd.to_datetime(df['order_time'],format='%H:%M:%S')

df['order_hour'] = df['order_time'].dt.hour

orders_by_hour = df.groupby('order_hour', observed= False)['order_id'].nunique()

ax = orders_by_hour.plot(kind='bar', color='maroon', figsize=(10,6), edgecolor='black')

plt.xlabel('Hours')
plt.ylabel('No of orders')

for i,j in enumerate(orders_by_hour):
    plt.text(i,j+1, str(j), ha = 'center', va = 'bottom', fontsize=12, fontstyle='italic')

plt.tight_layout()
plt.show()
```

```python
#Monthly Trend - total orders

df['order_date'] = pd.to_datetime(df['order_date'],dayfirst=True)

df['order_month'] = df['order_date'].dt.month_name()

month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

df['order_month'] = pd.Categorical(df['order_month'], categories=month_order, ordered = True)

grouporders_by_month = df.groupby('order_month', observed = False)['order_id'].nunique()

plot = grouporders_by_month.plot(kind = 'bar', color  = 'violet', figsize = (10,7), edgecolor = 'Yellow')

plt.xlabel('Months')
plt.ylabel('Total Orders')
plt.title('Total Orders by Month', fontsize = 20)

for i,j in enumerate(grouporders_by_month):
    plt.text(i,j+1,f'{j}', ha = 'center', va = 'bottom', fontsize=12, fontstyle='italic')

plt.tight_layout()
plt.show()
```

```python
#Monthly Trend - total orders

df['order_date'] = pd.to_datetime(df['order_date'],dayfirst=True)

df['order_month'] = df['order_date'].dt.month_name()

month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

df['order_month'] = pd.Categorical(df['order_month'], categories=month_order, ordered = True)

grouporders_by_month = df.groupby('order_month', observed = False)['order_id'].nunique()

plt.figure(figsize = (10,7))
plt.fill_between(grouporders_by_month.index, grouporders_by_month.values,color='yellow', alpha=0.7)
plt.plot(grouporders_by_month.index,grouporders_by_month.values, linewidth = 2, marker = 'o', color = 'red')

plt.xlabel('Months')
plt.ylabel('Total Orders')
plt.title('Total Orders by Month', fontsize = 20)

for i,j in enumerate(grouporders_by_month):
    plt.text(i,j+1,f'{j}', ha = 'center', va = 'bottom', fontsize=12, fontstyle='italic')

plt.tight_layout()
plt.show()
```

```python
## % Sales by category
category_sales = df.groupby('pizza_category')['total_price'].sum()
Percentofsales = category_sales / category_sales.sum()* 100

plt.figure(figsize = (5,7))
colors= plt.get_cmap('tab20').colors
plt.pie(Percentofsales, labels=Percentofsales.index, autopct='%1.1f%%', startangle=70, colors=colors )
centre_circle = plt.Circle((0,0), 0.70, fc='white')  # to make it a donut chart
plt.gca().add_artist(centre_circle) # to add the white circle in the middle
plt.title('Sales by Category', fontsize = 14)
plt.show()
```

```python
## sales by pizza size & category


sales_pivot = df.pivot_table(
    index = 'pizza_category',
    columns = 'pizza_size',
    values='total_price',
    aggfunc='sum',
    fill_value=0,
)

sales_pct = sales_pivot / sales_pivot.sum().sum() * 100

plt.figure(figsize=(10,6))
sns.heatmap(sales_pct, annot = True, fmt = '.2f', cmap = 'YlGnBu', linewidth = 0.5)
plt.show()
```

```python
## Total Pizzas sold by pizza catrgory 

pizza_by_category = df.groupby('pizza_category')['quantity'].sum()

colors = list(plt.get_cmap('tab20').colors)

colors = colors[:len(pizza_by_category)]

ax = pizza_by_category.plot(kind = 'bar', figsize=(10,6), color= colors, edgecolor='black')

plt.title("Total Pizzas Sold by Category", fontsize = 20)
plt.xlabel("Pizza Category", fontsize = 14) 
plt.ylabel("Number of Pizzas Sold", fontsize = 14)

for i,j in enumerate(pizza_by_category):
    plt.text(i, j+1, str(j), ha='center', fontsize=12, va = 'bottom', fontstyle='italic')
plt.tight_layout()
plt.show()
```

```python
# top 5 best selling pizzas

pizza_5_name = df.groupby('pizza_name')['quantity'].sum()

top5 = pizza_5_name.sort_values(ascending = False).head(5)

ax= top5.plot(kind = 'bar', figsize=(10,6), color='orange', edgecolor='black')

plt.title("top 5 best selling pizzas", fontsize = 20)
plt.xlabel("Pizza name")
plt.ylabel("No of pizzas sold")

for i,j in enumerate(top5):
    plt.text(i,j+1,str(j), ha= 'center', va= 'bottom', fontsize = 12, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
## bottom 5 pizzas - by sales 

pizza_5_name = df.groupby('pizza_name')['total_price'].sum()

top5 = pizza_5_name.sort_values(ascending = True).head(5)

ax= top5.plot(kind = 'bar', figsize=(10,6), color='green', edgecolor='black')

plt.title("top 5 best selling pizzas", fontsize = 20)
plt.xlabel("Pizza name")
plt.ylabel("No of pizzas sold")

for i,j in enumerate(top5):
    plt.text(i,j+1,str(j), ha= 'center', va= 'bottom', fontsize = 12, fontweight='bold')

plt.tight_layout()
plt.show()
```
